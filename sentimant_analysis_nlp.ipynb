```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('train.txt',sep=';',names=['text','emotion'])
df.head()
```

```python
unique_emotion=df['emotion'].unique()
```

```python
emotion_numbers={}
i=0
for emo in unique_emotion:
  emotion_numbers[emo]=i
  i=i+1
emotion_numbers
```

```python
df['emotion']=df['emotion'].map(emotion_numbers)
```

```python
df
```

```python
df['text']=df['text'].apply(lambda x : x.lower())
```

```python
#removing punctuation
import string
def remove_pun(text):
  return text.translate(str.maketrans('','',string.punctuation))
```

```python
df['text']=df['text'].apply(remove_pun)
```

```python
def remove_numbers(txt):
    new = ""
    for i in txt:
        if not i.isdigit():
            new = new + i
    return new

df['text'] = df['text'].apply(remove_numbers)
```

```python
def remove_emojis(txt):
    new = ""
    for i in txt:
        if i.isascii():
            new += i
    return new

df['text'] = df['text'].apply(remove_emojis)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def remove(txt):
  words = word_tokenize(txt)  #txt.spli()
  cleaned = []
  for i in words:
    if not i in stop_words:
      cleaned.append(i)

  return ' '.join(cleaned)
```

```python
import nltk
nltk.download('punkt_tab')
df['text']=df['text'].apply(remove)
```

```python
# bag of word technique of vetorization

from sklearn.feature_extraction.text import CountVectorizer

document=['i love pizza','i love coldrink','i love burger']

vectorizer=CountVectorizer(ngram_range=(1,2))
X=vectorizer.fit_transform(document)
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['text'],df['emotion'],test_size=0.2,random_state=42)

from sklearn.feature_extraction.text import TfidfVectorizer , CountVectorizer
vectorizer=TfidfVectorizer()
X_train_idfd=vectorizer.fit_transform(X_train)
X_test_idfd=vectorizer.fit_transform(X_test)

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
model=MultinomialNB()
model.fit(X_train_idfd,y_train)
```

```python
#  tfidf with naive baies
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

X_train_idfd = vectorizer.fit_transform(X_train)
X_test_idfd = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_idfd, y_train)

ypred = model.predict(X_test_idfd)
print(accuracy_score(y_test,ypred))
```

```python
# bag of words with countvectorizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
vectorizer = CountVectorizer()
X_train_bow = vectorizer.fit_transform(X_train)
X_test_bow = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_bow, y_train)

ypred = model.predict(X_test_bow)
print(accuracy_score(y_test,ypred))
```

```python
# tfidf with logistic regression
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.2, random_state=42)
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

model = LogisticRegression()
model.fit(X_train_tfidf, y_train)

ypred = model.predict(X_test_tfidf)
print("Accuracy:", accuracy_score(y_test, ypred))
```
