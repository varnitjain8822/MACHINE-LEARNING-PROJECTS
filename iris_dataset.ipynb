```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
x=df.drop('species',axis=1)
y=df['species']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn=KNeighborsClassifier()
model_knn.fit(x_train,y_train)
```

```python
model_knn.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
model_svm=SVC(C=30,gamma='auto',kernel='rbf')
model_svm.fit(x_train,y_train)
```

```python
model_svm.score(x_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
classifier=GridSearchCV(estimator=model_svm, param_grid={'C':[10,20,30],'kernel':['rbf','linear']}, cv=5,return_train_score=False)
classifier.fit(x_train,y_train)
```

```python
classifier.cv_results_
```

```python
df=pd.DataFrame(classifier.cv_results_)
df
```

```python
from sklearn.model_selection import RandomizedSearchCV
classifier=RandomizedSearchCV(model_svm, {'C':[10,20,30],'kernel':['rbf','linear']}, cv=5,return_train_score=False)
classifier.fit(x_train,y_train)
```

```python
df=pd.DataFrame(classifier.cv_results_)
df
```
